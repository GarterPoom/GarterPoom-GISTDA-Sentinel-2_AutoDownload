```python
import folium
from shapely.geometry import Polygon, mapping

# Define the polygon coordinates
coords = [(92.0, 28.5), (109.5, 28.5), (109.5, 5.5), (92.0, 5.5), (92.0, 28.5)]

# Create a map centered around the polygon
m = folium.Map(location=[17.0, 100.0], zoom_start=5)

# Create a polygon and add it to the map
polygon = Polygon(coords)
geojson = mapping(polygon)

folium.GeoJson(geojson, style_function=lambda x: {'fillColor': 'blue', 'color': 'black', 'weight': 2, 'fillOpacity': 0.4}).add_to(m)

# Display the map
m
```
